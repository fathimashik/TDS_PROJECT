```python
from sklearn.model_selection import train_test_split

# Split data (20% test, 80% train)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)
y_train.value_counts()
y_train.value_counts().idxmin()
```

```python
import numpy as np

X_train = X_train.replace("Unknown", np.nan)
X_test = X_test.replace("Unknown", np.nan)
age_mean = X_train["Age"].mean()
difficulty_mode = X_train["GameDifficulty"].mode()[0]
X_train["Age"].fillna(age_mean, inplace=True)
X_test["Age"].fillna(age_mean, inplace=True)
X_train["Location"].fillna("Other", inplace=True)
X_test["Location"].fillna("Other", inplace=True)
X_train["GameDifficulty"].fillna(difficulty_mode, inplace=True)
X_test["GameDifficulty"].fillna(difficulty_mode, inplace=True)
X_train["InGamePurchases"].fillna('0', inplace=True)
X_test["InGamePurchases"].fillna('0', inplace=True)
age_sum_test = round(X_test["Age"].sum(), 2)

print(age_sum_test)


```

```python
X_train = X_train.drop("PlayerID", axis=1)
X_test = X_test.drop("PlayerID", axis=1)
difficulty_mapping = {"Easy": 0, "Medium": 1, "Hard": 2}

X_train["GameDifficulty"] = X_train["GameDifficulty"].map(difficulty_mapping)
X_test["GameDifficulty"] = X_test["GameDifficulty"].map(difficulty_mapping)
X_train = pd.get_dummies(
    X_train,
    columns=["Gender", "Location", "GameGenre"],
    drop_first=True
)

X_test = pd.get_dummies(
    X_test,
    columns=["Gender", "Location", "GameGenre"],
    drop_first=True
)
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

import numpy as np

total_sum = round(np.sum(X_test_scaled[:5]), 2)

print(total_sum)

```
