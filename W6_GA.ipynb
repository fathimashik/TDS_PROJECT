```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load dataset
wine = load_wine()
X = wine.data
y = wine.target

# Train-test split (70:30, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)

# Train Logistic Regression model
model = LogisticRegression(
    max_iter=10000,
    C=0.5,
    random_state=1,
    solver='sag'
)
model.fit(X_train, y_train)

# Predict on test data
y_pred = model.predict(X_test)

# Compute accuracy
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
print(model.n_iter_)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define model
dt = DecisionTreeClassifier(random_state=1)

# Parameter grid
param_grid = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'min_samples_leaf': [2, 4, 6, 8, 10],
    'max_depth': [3, 4, 5, 6]
}

# GridSearchCV
grid = GridSearchCV(dt, param_grid=param_grid, cv=4)
grid.fit(X_train, y_train)

# Best model
best_model = grid.best_estimator_

# Predict and evaluate
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(accuracy)
```

```python
print(grid.best_params_['max_depth'])
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define model
ada = AdaBoostClassifier(random_state=1)

# Parameter grid
param_grid = {
    'n_estimators': [100, 500, 1000],
    'learning_rate': [0.5, 1, 2]
}

# GridSearchCV
grid = GridSearchCV(ada, param_grid=param_grid, cv=4)
grid.fit(X_train, y_train)

# Best model
best_model = grid.best_estimator_

# Predict and compute accuracy
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(accuracy)
```

```python
print(grid.best_params_['n_estimators'])
```

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.metrics import accuracy_score

# Define base models
bag = BaggingClassifier(random_state=1)
rf = RandomForestClassifier(random_state=1)
gb = GradientBoostingClassifier(random_state=1)

# Voting Classifier
voting = VotingClassifier(
    estimators=[('bag', bag), ('rf', rf), ('gb', gb)],
    voting='hard'
)

# Train
voting.fit(X_train, y_train)

# Predict
y_pred = voting.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
